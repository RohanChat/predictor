```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


base_url = "http://www.ufcstats.com/statistics/events/completed?page="


all_events = []


for page_num in range(1, 29):
   
    url = f"{base_url}{page_num}"
    
   
    response = requests.get(url)
    
   
    soup = BeautifulSoup(response.text, 'html.parser')
    
    
    for section in soup.find_all('tbody'):
        
        for row in section.find_all('tr', class_='b-statistics__table-row'):
            
            event_name = None
            event_link = None
            event_location = None
            event_date = None
            
            
            event_name_tag = row.find('a', class_='b-link b-link_style_black')
            if event_name_tag:
                event_name = event_name_tag.text.strip()
                event_link = event_name_tag['href']
            
            
            event_location_tag = row.find('td', class_='b-statistics__table-col_style_big-top-padding')
            if event_location_tag:
                event_location = event_location_tag.text.strip()
                
            # Extract the event date
            event_date_tag = row.find('span', class_='b-statistics__date')
            if event_date_tag:
                event_date = event_date_tag.text.strip()
            
            # Only append rows that have an event name, link, location, and date
            if event_name and event_link and event_location and event_date:
                all_events.append([event_name, event_link, event_location, event_date])


df = pd.DataFrame(all_events, columns=['Event Name', 'Event Link', 'Location', 'Date'])

print(df.head())
df.to_csv('ufc_events.csv', index=False)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


events_df = pd.read_csv('ufc_events.csv')


event_names = []
event_links = []
locations = []
dates = []
fight_links = []
fighters_B = []
fighters_R = []
winners = []
fighter_B_KD = []
fighter_R_KD = []
fighter_B_STR = []
fighter_R_STR = []
fighter_B_TD = []
fighter_R_TD = []
fighter_B_SUB = []
fighter_R_SUB = []
weight_classes = []
methods = []
rounds = []
times = []


for index, event in events_df.iterrows():
    event_name = event['Event Name']
    event_link = event['Event Link']
    location = event['Location']
    date = event['Date']
    
    
    response = requests.get(event_link)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    
    table = soup.find('tbody', class_='b-fight-details__table-body')
    
    
    if not table:
        continue
    
    
    for row in table.find_all('tr'):
        
        fight_link = row['data-link']
        
        
        win_column = row.find('a', class_='b-flag_style_green')
        winner = 'Red' if win_column else 'Blue'
        
        
        fighters = row.find_all('a', class_='b-link_style_black')
        fighter_R = fighters[0].get_text(strip=True)
        fighter_B = fighters[1].get_text(strip=True)
        
        
        fight_stats = row.find_all('td', class_='b-fight-details__table-col')
        
        kd_r = fight_stats[2].find_all('p')[0].get_text(strip=True)  # Red fighter KD
        kd_b = fight_stats[2].find_all('p')[1].get_text(strip=True)  # Blue fighter KD
        
        str_r = fight_stats[3].find_all('p')[0].get_text(strip=True)  # Red fighter STR
        str_b = fight_stats[3].find_all('p')[1].get_text(strip=True)  # Blue fighter STR
        
        td_r = fight_stats[4].find_all('p')[0].get_text(strip=True)  # Red fighter TD
        td_b = fight_stats[4].find_all('p')[1].get_text(strip=True)  # Blue fighter TD
        
        sub_r = fight_stats[5].find_all('p')[0].get_text(strip=True)  # Red fighter SUB
        sub_b = fight_stats[5].find_all('p')[1].get_text(strip=True)  # Blue fighter SUB
        
        weight_class = fight_stats[6].find('p').get_text(strip=True)  # Weight Class
        method = fight_stats[7].find('p').get_text(strip=True)  # Method
        round_fought = fight_stats[8].find('p').get_text(strip=True)  # Round
        time_fought = fight_stats[9].find('p').get_text(strip=True)  # Time
        
        
        event_names.append(event_name)
        event_links.append(event_link)
        locations.append(location)
        dates.append(date)
        fight_links.append(fight_link)
        fighters_R.append(fighter_R)
        fighters_B.append(fighter_B)
        winners.append(winner)
        fighter_B_KD.append(kd_b)
        fighter_R_KD.append(kd_r)
        fighter_B_STR.append(str_b)
        fighter_R_STR.append(str_r)
        fighter_B_TD.append(td_b)
        fighter_R_TD.append(td_r)
        fighter_B_SUB.append(sub_b)
        fighter_R_SUB.append(sub_r)
        weight_classes.append(weight_class)
        methods.append(method)
        rounds.append(round_fought)
        times.append(time_fought)


fight_df = pd.DataFrame({
    'Event Name': event_names,
    'Event Link': event_links,
    'Location': locations,
    'Date': dates,
    'Fight Link': fight_links,
    'Fighter_R': fighters_R,
    'Fighter_B': fighters_B,
    'Winner': winners,
    'Fighter_B_KD': fighter_B_KD,
    'Fighter_R_KD': fighter_R_KD,
    'Fighter_B_STR': fighter_B_STR,
    'Fighter_R_STR': fighter_R_STR,
    'Fighter_B_TD': fighter_B_TD,
    'Fighter_R_TD': fighter_R_TD,
    'Fighter_B_SUB': fighter_B_SUB,
    'Fighter_R_SUB': fighter_R_SUB,
    'Weight Class': weight_classes,
    'Method': methods,
    'Round': rounds,
    'Time': times
})


fight_df.to_csv('all_fightrs.csv', index=False)
print(fight_df)


```
